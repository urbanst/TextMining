```python
# Now we can just re-load the already trained model
d2v_model= Doc2Vec.load("Doc2Vec/Doc2Vec.model")
```

```python
#This extracts the tags and their vectors from the trained model.
doc_tags = list(d2v_model.docvecs.doctags.keys())
X = d2v_model[doc_tags]

#This t-SNE reduces our trained vectors (i.e., 100-dimensional) into 2-dimensional vectors, and we store the reduced vectors into a pandas dataframe.
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
df = pd.DataFrame(X_tsne, index=doc_tags, columns=['x', 'y'])

# let's see transformed vectors into 2d dimensional space with tsne to print into diagram
print("20 dimensional vectors of document transformed into 2D array.")
print(df[:10])# data frame consists of document tags with vectors transformed in 2 dimensial xpace (x,y)
```
